#### 설명
① 3번 문제에서 확인한 결과에 따르면, A와 B는 chroma cqt 의 진행 흐름이 유사한 것을 확인할 수 있습니다. 하지만 이 정도 결과만으로 'A는 C보다 B랑 가까워' 라고 말하기는 쉽지 않습니다.

우리가 무언가를 멀다 혹은 가깝다라고 이야기 하기 위해서는 거리 개념이 정의되어야 합니다. 앞선 예제에서 우리가 추출한 chroma_cqt 는 [12, time] 의 크기를 갖는 행렬이며, 이러한 행렬 두개 사이에서 거리를 정의할 수 있는 적절한 방법이 있어야만, A-B 와 A-C 사이의 거리의 크고 작음을 판단할 수 있습니다.

② 가장 간단한 방법으로는 두 행렬의 각 원소들 사이의 euclidean distance를 계산하는 것이지만, spectrogram을 살펴보면 이를 바로 적용하기 어렵다는 점을 확인할 수 있습니다. 두 곡의 key가 다르기 때문이죠.

1번 예제에서 음원을 통해 들으셨던 대로 A곡과 B곡은 서로 다른 키로 진행되고 있습니다. MIR에서는 이러한 키 차이를 간단하게 보정할 수 있는 방법이 존재하는데, 그 방법이 바로 Optimal transposition index를 활용한 방법입니다. [[1]](http://mtg.upf.edu/files/publications/jserra08TransposigChroma2commonKey.pdf)

③ 논문에서 제시된 내용을 현재 상황에 맞추면, OTI를 계산하는 방법은 아래 순서와 같습니다.

1) 비교할 두 곡의 chroma cqt $$A, B \in R^{12 \times T}$$ 를 계산한다.

2) A와 B 를 time축에 대해 평균을 취한 $$g_A, g_B$$ 를 얻는다.

3) g_A를 고정한 채, g_B를 12번 circular shift 하며 두 벡터 간 내적을 구한다.

4) 내적 값이 가장 큰 circular shift index를 반환한다.

즉, 주어진 곡을 12번 전조 시키면서, 비교 대상 곡과 가장 유사한 순간의 전조 횟수가 얼마인지를 계산하는 방법입니다.
